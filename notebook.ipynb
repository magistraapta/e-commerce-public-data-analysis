### **Cleaning Data**

delete duplicates from geolocation datasets

```python
geolocation.duplicated().sum()
```

```python
geolocation.drop_duplicates(inplace=True)
geolocation.duplicated().sum()
```

Change the datetime data type from objc to date-time

```python
orders_col = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
    ]

for col in orders_col:
    orders[col] = pd.to_datetime(orders[col])

orders.info()
```

```python
order_reviews_col = [
    'review_creation_date',
    'review_answer_timestamp'
]

for col in order_reviews_col:
    order_reviews_df[col] = pd.to_datetime(order_reviews_df[col])

order_reviews_df.info()
```

```python
order_items_col = ['shipping_limit_date']

for col in order_items_col:
    order_items_df[col] = pd.to_datetime(order_items_df[col])
    
order_items_df.info()
```

Handle the missing value

```python
## change the missing value for title and message review
order_reviews_df['review_comment_title'].fillna('No Title', inplace=True)
order_reviews_df['review_comment_message'].fillna('No Message', inplace=True)
```

```python
order_reviews_df.head(5)
```

Drop unused cols on product table

```python
product_drop_cols = [
    'product_name_lenght',
    'product_description_lenght',
    'product_weight_g',
    'product_length_cm',
    'product_height_cm',
    'product_width_cm'
]

product.drop(product_drop_cols, axis=1, inplace=True)
```

## **Exploratory Data Analysis**

Explore customer & order

```python
customer_order_df = pd.merge(
    left=customer,
    right=orders,
    how='left',
    left_on='customer_id',
    right_on='customer_id'
)

customer_order_df.head(5)
```

Explore payment & reviews

```python
payments_reviews_df = pd.merge(
    left=order_payments_df,
    right=order_reviews_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)

payments_reviews_df.head(5)
```

```python
payments_reviews_df.sort_values(by='payment_value', ascending=False)
```

```python
customer_df = pd.merge(
    left=customer_order_df,
    right=payments_reviews_df,
    how='left',
    left_on='order_id',
    right_on='order_id'
)

customer_df.head(5)
```

Explore item & seller

```python
seller_item_df = pd.merge(
    left=order_items_df,
    right=sellers,
    how='left',
    left_on='seller_id',
    right_on='seller_id'
)

seller_item_df.head(5)
```

Explore product & category

```python
product_category_df = pd.merge(
    left=product,
    right=category,
    how='left',
    left_on='product_category_name',
    right_on='product_category_name'
)

product_category_df.head(5)
```

```python
product_category_df.groupby(by='product_category_name_english').product_id.nunique().sort_values(ascending=False).head(10)
```

Explore seller & product_category

```python
seller_product_df = pd.merge(
    left=product_category_df,
    right=seller_item_df,
    how="left",
    left_on="product_id",
    right_on="product_id"
)
seller_product_df.head()
```

```python
all_df = pd.merge(
    left=customer_df,
    right=seller_product_df,
    how='left',
    left_on='order_id',
    right_on='order_id'
)

all_df.head()
```

```python
all_df = all_df.drop_duplicates('order_id')
```

```python
all_df.groupby(by='product_category_name_english').agg({
    'order_id': 'nunique',
    'review_score': ['min', 'max']
})
```

```python
product_count = all_df.groupby('product_category_name_english').product_id.count().reset_index()
sorted_product = product_count.sort_values(by='product_id', ascending=False)

sorted_product
```

```python
all_df.to_csv('all_data.csv', index=False)
```

## **Data Visualizations**

### What is the highest and lowest selling product?

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(x="product_id", y="product_category_name_english", data=sorted_product.head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("products with the highest sales", loc="center", fontsize=18)
ax[0].tick_params(axis ='y', labelsize=15)

sns.barplot(x="product_id", y="product_category_name_english", data=sorted_product.sort_values(by="product_id", ascending=True).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("products with the lowest sales", loc="center", fontsize=18)
ax[1].tick_params(axis='y', labelsize=15)
plt.show()
```

```python
delivery_status = all_df['order_status'].value_counts().sort_values(ascending=False)
delivery_status
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=delivery_status.index, 
    y=delivery_status.values, 
    order=delivery_status.index,
    palette=colors
)

plt.title("Successful Delivery", fontsize=15)
plt.xlabel("Delivery status")
plt.ylabel("Count")
plt.xticks(fontsize=12)
plt.show()
```
